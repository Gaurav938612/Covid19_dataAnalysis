```python
data = client["covidDB"]
```

```python
#Creating Collections
age_group_details_db   =data["age_group_details_db"]           
covid_india_db         =data["covid_india_db"]         
hospital_beds_india_db =data["hospital_beds_india_db"]         
icmr_testing_details_db=data["icmr_testing_details_db"]        
icmr_testing_labs_db   =data["icmr_testing_labs_db"]         
individual_details_db  =data["individual_details_db"]         
population_india_db    =data["population_india_db"]         
statewise_testing_db   =data["statewise_testing_db"]   
```

```python
import json
#saving the age_group_details data to a json file
age_group_details.to_json('age_details.json')
```

```python
#loading the saved json file 
file=open('age_details.json').read()
loaded_data=json.loads(file)
```

```python
def converter(d):
    columns=[]
    for i in d.keys():
        columns.append(i)
    
    start=columns[0]
    length=len(d[start].keys())
    #print(length)
    ans=[]
    for i in range(length):
        temp={}
        for j in columns:
            temp[j]=d[j][str(i)]
        ans.append(temp)
    return ans
```

```python
a=age_group_details_db.insert_many(converter(loaded_data))
print(a.inserted_ids)
```

```python
#query  : find all records of age groups which comprises more than 10% in corona cases
#solution:
query= { "Percentage": { "$gt": 10 } }
document = age_group_details_db.find(query)

for i in document:
  print(i) 
```

```python
#query  : find all the records of age groups having more than 100 cornoa cases
#solution:
query={"TotalCases":{"$gt":100}}
document=age_group_details_db.find(query)
for i in document:
  print(i)
```
